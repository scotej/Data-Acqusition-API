# Gathering data using an API

## Step 1: Get an API Key

Go to [OpenWeatherMap](https://openweathermap.org/appidand) sign up for a free account.

Once registered, navigate to the API keys section to generate your API key.

Copy the API key—you’ll need it to make requests.

## Step 2: Install Required Libraries

To interact with the API, you’ll need the requests library in Python. Install it using pip:

`pip install requests`

## Step 3: Make Your First API Request

Let’s start by fetching the current weather for a specific city.

```python
import requests
import json

# Replace with your API key
API_KEY = 'REDACTED_API_KEY'

# API endpoint and parameters
city = 'Melbourne'
url = f'http://api.openweathermap.org/data/2.5/weather?q={city}&appid={API_KEY}&units=metric'

# Make the request
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()  # Parse JSON response
    pretty_json_string = json.dumps(data, indent=4)
    print(pretty_json_string)  # Print the raw data
else:
    print(f"Error: {response.status_code}")
```

## Step 4: Parse and Use the Data

The API returns data in JSON format. Let’s extract and display some useful information:

```python
# Extract relevant data
if response.status_code == 200:
    data = response.json()
    city_name = data['name']
    weather_description = data['weather'][0]['description']
    temperature = data['main']['temp']  # Temperature in Kelvin
    humidity = data['main']['humidity']

    # Convert temperature to Celsius
    temperature_celsius = temperature

    # Display the data
    print(f"Weather in {city_name}:")
    print(f"Description: {weather_description}")
    print(f"Temperature: {temperature_celsius:.2f}°C")
    print(f"Humidity: {humidity}%")
```

**Task 5**: What units does the API use for temperature, wind speed, and other measurements by default? Can you change these units?

<span style="color:red">(1 mark)</span>


The OpenWeatherMap API supports three unit systems:

### 1. Default (no units parameter):
- Temperature: Kelvin (K)
- Wind Speed: meters/second (m/s)
- Pressure: hPa (hectopascal)
- Humidity: % (percentage)

### 2. Metric (add `&units=metric` to URL):
- Temperature: Celsius (°C)
- Wind Speed: meters/second (m/s)
- Pressure: hPa (hectopascal)
- Humidity: % (percentage)

### 3. Imperial (add `&units=imperial` to URL):
- Temperature: Fahrenheit (°F)
- Wind Speed: miles/hour (mph)
- Pressure: hPa (hectopascal)
- Humidity: % (percentage)

```python
import requests
import json

API_KEY = 'REDACTED_API_KEY'
city = 'Beijing'

# Kelvin
print("DEFAULT UNITS (Kelvin):")
url_default = f'http://api.openweathermap.org/data/2.5/weather?q={city}&appid={API_KEY}'
response = requests.get(url_default)
if response.status_code == 200:
    data = response.json()
    print(f"Temperature: {data['main']['temp']} K")
    print(f"Wind Speed: {data['wind']['speed']} m/s")

# Metric
print("METRIC UNITS (Celsius):")
url_metric = f'http://api.openweathermap.org/data/2.5/weather?q={city}&appid={API_KEY}&units=metric'
response = requests.get(url_metric)
if response.status_code == 200:
    data = response.json()
    print(f"Temperature: {data['main']['temp']} °C")
    print(f"Wind Speed: {data['wind']['speed']} m/s")

#Imperial
print("IMPERIAL UNITS (Fahrenheit):")
url_imperial = f'http://api.openweathermap.org/data/2.5/weather?q={city}&appid={API_KEY}&units=imperial'
response = requests.get(url_imperial)
if response.status_code == 200:
    data = response.json()
    print(f"Temperature: {data['main']['temp']} °F")
    print(f"Wind Speed: {data['wind']['speed']} mph")

```
